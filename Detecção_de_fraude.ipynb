```python
# Seleciona colunas num√©ricas
num_cols = dados.select_dtypes(include=np.number).columns.tolist()

# Fun√ß√£o para m√©tricas customizadas
def calc_custom_stats(df):
    return pd.DataFrame({
        "q01": df.quantile(0.01),
        "q05": df.quantile(0.05),
        "q10": df.quantile(0.10),
        "q25": df.quantile(0.25),
        "q75": df.quantile(0.75),
        "q90": df.quantile(0.90),
        "q95": df.quantile(0.95),
        "q99": df.quantile(0.99),
        "iqr": df.quantile(0.75) - df.quantile(0.25),
        "mad": df.apply(lambda x: np.median(np.abs(x - np.median(x))))
    })

# Estat√≠sticas globais
summary_global = pd.concat([
    dados[num_cols].agg(["count","mean","std","min","median","max"]).T,
    calc_custom_stats(dados[num_cols])
], axis=1)

# Estat√≠sticas por classe
grouped = dados.groupby("Class")[num_cols]

summary_by_class_basic = grouped.agg(["count","mean","std","min","median","max"])

# Estat√≠sticas customizadas por classe
summary_by_class_custom = grouped.apply(calc_custom_stats)
summary_by_class_custom = summary_by_class_custom.unstack(level=1)
summary_by_class_custom.columns = [f"{stat}_{col}" for col, stat in summary_by_class_custom.columns]

# Concatenar m√©tricas
summary_by_class = pd.concat([summary_by_class_basic, summary_by_class_custom], axis=1)

# Formata√ß√£o para melhor legibilidade
pd.options.display.float_format = "{:,.4f}".format

# Exibir resultados
display(summary_global)

print("\n Estatisticas por Classes\n")
display(summary_by_class)
```

## <font color = "red">Teste Estatistico</font>

Utilizamos testes estat√≠sticos com o objetivo de analisar as vari√°veis envolvidas e formular hip√≥teses relevantes. Nesta etapa do estudo, foram aplicados os seguintes testes:



1.  <font color = "orange"> ***Mann-Whitney U Test(Wilcoxon Rank-Sum Test)***</font> :

<font color = "grey"> **O que √©**</font>
*   √â um teste estat√≠stico n√£o param√©tricos usado para comparar duas amostras independentes.

*   Pergunta : "as duas amostras v√™m da mesma distribui√ß√£o

*   √â usado quando os dados n√£o s√£o normais  ou quando n√£o queremos assumir que a m√©dia √© o suficiente

<font color = "grey"> **Quando usar?!**</font>

*   Comparar vari√°veis cont√≠nuas entre dois grupos especialmente quando a normalidade n√£o √© garantida

*   Exemplo: Comparando valores de "Amount" entre fraudes e n√£o fraudes.

[Documenta√ß√£o Mann-Whitney U Test](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mannwhitneyu.html)

2.   <font color = "orange"> ***Point-Biserial Correlation*** </font> :

<font color = "grey"> **O que √©**</font> :

*   E uma correla√ß√£o entre uma vari√°vel bin√°ria e uma cont√≠nua;

*   Pergunta : "qu√£o linearmente relacionada est√° a vari√°vel cont√≠nua com a vari√°vel bin√°ria?"

*   √â equivalente ao <font color = "green"> Pearson correlation</font>, mas aplicado a uma vari√°vel bin√°ria.

<font color = "grey"> **Quando usar**</font> :

*   Avaliar associa√ß√£o entre uma feature cont√≠nua e o alvo bin√°rio(neste caso a *Class*)

*   Exemplo: verificar se *Amount* tem rela√ß√£o com *Class*(fraude ou n√£o)

[Documenta√ß√£o Pont-Biserial Correlation](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pointbiserialr.html)

3.   <font color = "orange">***FDR-False Discovery Rate(Corre√ß√£o de m√∫ltiplos testes)***</font> :

<font color = "grey"> **O que √©**</font> :

*   Quando voc√™ faz muitos testes estat√≠sticos, aumenta a chance de encontrar falsos positivos(erro tipo 1)

*   <font color = "orange">**FDR**</font> √© uma t√©cnica que ajusta os p-values para controlar esses falsos positivos

<font color = "grey"> **Quando usar**</font> :

*   Sempre que voc√™ fizer muito testes simultaneamente

*   exemplo: testar 30 colunas num√©ricas pra rela√ß√£o com <font color = "orange">**Class**</font>

üìö<font color = "grey"> **Leituras Recomendadas**</font> :

<font color = "green"> Artigo explicativo </font> :

[Benjamini-Hochberg procedure- Wikipedia](https://en.wikipedia.org/wiki/False_discovery_rate#Benjamini%E2%80%93Hochberg_procedure)

[False Discovery Rate](https://www.publichealth.columbia.edu/research/population-health-methods/false-discovery-rate?utm_source=chatgpt.com)







4.   <font color = "orange">Subamostragem para acelerar</font> :

*   Quando o dataset √© grande, como neste que  cont√©m um pouco mais de 280 mil linhas, testes estat√≠sticos podem ser lentos;

*   **Subamostragem**: pegar apenas uma amostra aleat√≥ria menor(ex:5000 linhas), mantendo a representatividade

*   **Por que funciona**: <font color = "orange">Mann-Whitney</font> e <font color = "orange">correla√ßao ponto biserial</font> dependem das distribui√ßoes e rela√ß√µes gerais, ent√£o amostras grandes ainda representam bem os dados.


```python
#Pegando colunas de tipo numericos, armazenando todas as colunas desta caracteristica como uma lista
n_cols = dados.select_dtypes(include="number").columns.tolist()

# removendo a variavel "Class", para n√£o comparar com ela mesmo
n_cols = [c for c in n_cols if c != "Class"]

# lista vazia para armazenagem dos dados
rows = []

# criando mascara booleana
fraud = dados["Class"] == 1

# criando um loop para pegar os valores
for col in n_cols:
  # pegando apenas valores com transa√ßoes fraudulentas
  a = dados.loc[fraud, col]
  # pegando apenas valores de transa√ß√µes nao fraudulentas
  b = dados.loc[~fraud, col]

  #pegando subamostragem para a velocidade
  # se houver mais de 15000, utilizaremos apenas uma amostra
  #random_state √© para garantir a reprodutibilidade
  a_s = a.sample(n = min(15000, len(a)), random_state= 70) if len(a.dropna()) > 15000 else a
  b_s = b.sample(n = min(15000, len(b)), random_state = 70) if len(b.dropna()) > 15000 else b

#Teste Mann-whitney
  try:
    #mannwhitneyu compara se as distribui√ßoes de a_s e b_s s√£o diferentes
    # two-sided teste bilateral(verificando qualquer diferen√ßa)
    mw_stat, mw_p = mannwhitneyu(a_s, b_s, alternative = "two_sided")

  except:
    mw_stat, mw_p = np.nan, np.nan
# Correla√ß√£o ponto-biseral
  try:
    #pointbiserialr mede a correla√ß√£o entre a variavel continua e bin√°ria(class)
    #pd_corr valor da correla√ßao
    #pb_p p-valor da correla√ßao
    pd_corr, pb_p = pointbiserialr(dados["Class"], dados[col])

  except:
    pd_corr, pb_p = np.nan, np.nan

# Armazendando os resultados
rows.append({"feature": col, "mw_p": mw_p, "pb_corr": pd_corr, "pb_p": pb_p})

#Transformando em DataFrame
tests = pd.DataFrame(rows)

# ajustando p-values oara multiplos testes
#multipletests corrige os p-values usando fdr, controlando erro tipo 1
tests["mw_p_adj"] = multipletests(tests["mw_p"].fillna(1.0), method="fdr_bh")[1]
tests["pb_p_adj"] = multipletests(tests["pb_p"].fillna(1.0), method="fdr_bh")[1]

# ordenar pelas correla√ß√µes mais forte
tests = tests.sort_values(by="pb_corr", key=lambda s: s.abs(), ascending=False)

tests.head()

```

OBS: O resultado do teste mw-p foi NaN, o provavel motivo desta resposta √© o desbalanceamento do teste. Como a amostra da classe positiva √© t√£o pequena(0.17%). foram poucos dados  para fazer as analises.

<font color = "orange"> ECDF(Empirical Cumulative Distribution Function)</font> :

<font color = "grey"> **O que √©**</font> :

√â a fun√ß√£o de distribui√ß√£o acumulada emp√≠rica.

*   Para um conjunto de dados, a ECDF mostra a propor√ß√£o de observa√ß√µes menorers ou iguais a um valor x.

*   Formalmente:

A **fun√ß√£o de distribui√ß√£o acumulada emp√≠rica (ECDF)** em um ponto $x$ √© dada por:

$$
ECDF(x) = \frac{\text{n√∫mero de observa√ß√µes } \leq x}{N}
$$

onde $N$ √© o total de observa√ß√µes na amostra.

<font color = "grey"> **Intui√ß√£o**</font> :

*   Imagine ordenar os valores do seu dataset;

*  A <font color = "orange">ECDF</font> vai de 0 a 1 e "sobe degraus" cada vez que um dado aparece;

*   Diferente de histogramas ou <font color = "orange"> KDE </font>, a <font color = "orange">ECDF</font> n√£o perde informa√ß√£o(√© uma representa√ß√£o exata da distribui√ß√£o dos dados).

<font color = "grey"> **Quando usar?**</font> :

*   Para visualizar distribui√ß√µes sem binning(diferente de histograma);

*   Para comparar distribui√ß√µes entre grupos(ex: fraude vs n√£o fraude);

*   Em estat√≠stica n√£o param√©trica, como base para testes de Kolmogorov-Smirnov

üìö<font color = "grey"> **Leituras Recomendadas**</font> :

[Documenta√ß√£o ECDF](https://www.statsmodels.org/stable/generated/statsmodels.distributions.empirical_distribution.ECDF.html)

[Histogram vs eCDF ‚Äì Marc‚Äôs Blog](https://brooker.co.za/blog/2022/09/02/ecdf.html?utm_source=chatgpt.com)






```python
#Chamando uma fun√ß√£o
def plot_ecdf_by_class(col):

  #Separando os dados por classes
  x0 = dados.loc[dados.Class == 0, col].dropna()
  x1 = dados.loc[dados.Class == 1, col].dropna()

  # Criando a fun√ß√£o de distribui√ß√£o acumulada
  ec0, ec1 = ECDF(x0), ECDF(x1)

  #definindo os eixos para a plotagem do gr√°fico
  #criando 400 pontos espa√ßados igualmente
  #entre o percentil 1% e 99%
  xs = np.linspace(
      np.percentile(dados[col], 1),
      np.percentile(dados[col], 99),
      400)

  #Plotando a curva ECDF
  plt.plot(xs, ec0(xs), label = "Class 0")
  plt.plot(xs, ec1(xs), label = "Class 1")

   #Customizando o gr√°fico, com titulo e nomes do eixos
  plt.title(f"ECDF de {col} por Classe", fontweight = "bold", fontsize = 20, pad = 20)
  plt.xlabel(col)
  plt.ylabel("F(x)")
  plt.legend()

  sns.despine()
  plt.show()
```

### <font color = "green">Balanceamento dos dados</font>

√â fundamental compreender as diferentes t√©cnicas de balanceamentos de dados, uma vez que a escolha adequada pode impactar significativamente o desempenho e a capacidade preditiva de um modelo de aprendizado de m√°quina. Quando os dados est√£o desbalanceados, ou seja, quando h√° uma distribui√ß√£o desigual entre as classes, o modelo tende a favorecer a classe majorit√°ria, comprometendo a acur√°cia em casos minorit√°rios.

Ecistem tr√™s abordagens principais para lidar com esse problema:


1.   <font color = "orange">Oversampling(Amostragem Excessiva)</font> :

Essa t√©cnica cria novas amostras da classe minorit√°ria para que ela tenha um n√∫mero de ocorr√™ncias semelhantes ao da classe majorit√°ria.

*   <font color = "orange">*SMOTE(Synthetic Minority Over-sampling Technique)*</font> : √â a t√©cnica mais popular. o  <font color = "orange">SMOTE</font> n√£o duplica amostras existentes; ele cria amostras sint√©ticas baseadas nas amostras vizinhas. isso √© feito encontrando os "vizinhos mais pr√≥ximos" de cada ponto da classe minorit√°ria e gerando novos pontos em algum lugar ao longo da linha que os conecta;

*    <font color = "orange">*Random Over-Sampling*</font> : √â a t√©cnica mais simples. Ela simplesmente duplica  aleatoriamente as amostras da classe minorit√°ria at√© que o conjunto de dados esteja balanceado. A desvantagem √© que pode levar ao <font color = "orange">overfitting</font>, pois o modelo pode memorizar as amostras duplicadas

üìö<font color = "grey"> **Leituras Recomendadas**</font> :

<font color = "green"> Artigo explicativo </font> :

[Artigo explicativo ](https://machinelearningmastery.com/random-oversampling-and-undersampling-for-imbalanced-classification/)

2.  <font color = "orange">Undersampling(Subamostragem)</font> :

Essa t√©cnica remove amostras da classe majorit√°ria para que ela se torne do mesmo tamanho da classe minorit√°ria.

*   <font color = "orange">*Random Undersampling*</font> : Remove aleatoriamente amostra de classe majorit√°ria. √â simples de implementar, mas a principal desvantagem √© que pode perder informa√ß√µes importantes ao descartar dados, o que pode levar a um desempenho ruim do modelo

*   <font color = "orange">*Tomek Links*</font> :  √â uma t√©cnica mais avan√ßada que identifica e remove pares de amostras(uma de cada classe) que s√£o os "vizinhos mais pr√≥ximos" um do outro, mas que s√£o de classes diferentes. A remo√ß√£o desses pares ajuda a melhorar a fronteira de decis√£o entre as classes.

3. <font color = "orange">Hibrido</font> :

Essa abordagem combina <font color = "orange">oversampling</font> e <font color = "orange">undersampling </font>  para otimizar o balanceamento

*   <font color = "orange">*SMOTE + Tomek Links*</font> : Essa √© uma combina√ß√£o popular. Primeiro, o <font color = "orange">SMOTE</font> √© aplicado para aumentar o n√∫meor de amostras da classe minorit√°ria. . Com isso, constru√≠mos um cen√°rio hipot√©tico de:

Quanto custar√° para a empresa o Falso Positivo (FP).

O custo da fraude efetiva (Falso Negativo - FN).

O quanto a empresa economizaria com as fraudes detectadas (Verdadeiro Positivo - VP).

Atrav√©s desta fun√ß√£o, calculamos tr√™s m√©tricas econ√¥mico-financeiras:

1. <font color = "darkpurple">ROI(Return on Investment) na fun√ß√£o de fraude</font>:

* ***O que √©***: Meded o retorno financeiro l√≠quido que voc√™ teve ao usar o modelo, comparando quanto voc√™ economizou evitando  fraudes com o custo de falsos positivos(alerta desnecess√°rios).

* F√≥rmula conceitual da fun√ß√£o:

<font color = "darkpurple">$$
ROI = \frac{(\text{VP} \times C_{\text{fraude}}) - (\text{FP} \times C_{\text{FP}})}{\text{FP} \times C_{\text{FP}}}
$$</font>,

onde;
   * VP = Verdadeiros Positivos(fraudes corretamenta detectadas)
   * FP = Falso Positivos(alertas incorretos)
   * $C_{\text{fraude}}$ = Custo de cada fraude
   * $C_{\text{FP}}$ = Custo de cada falso positivo.

Significado:
   
   * Se ROI > 0 -> modelo est√° "valendo a pena" financeiramente.
   * Se ROI < 0 -> custo com alertas falsos √© maior que o benef√≠cio de capturar fraudes





2. <font color = "darkpurple">Taxa de Economia</font>:

* ***O que √©***: percentual do custo total de fraude que o modelo conseguiu evitar.

* ***F√≥rmula conceitual***:

<font color = "darkpurple">$$
\text{Taxa de Economia} = \frac{VP \times C_{\text{fraude}}}{VP \times C_{\text{fraude}} + FN \times C_{\text{fraude}}} = \frac{VP}{VP + FN}
$$</font>

onde;
  * VP = Verdadeiro Positivo(fraudes verdadeiras detectadas)
  * $C_{\text{fraude}}$ = Custo de cada Fraude
  * FN = Falsos Negativos(fraudes que o modelo n√£o detectou)

Significado:
   * % quanto de prejuizo o modelo evitou com as fraudes.  


3. <font color = "darkpurple">Impacto L√≠quido</font>:

* ***O que √©***: √â valores desembolsado para cobrir fraudes e custo de campanha contra fraude  e o dinheiro recuperado quando o modelo evita a fraude.

* ***F√≥rmula conceitual***:
 <font color = "darkpurple">$$
\text{Impacto L√≠quido} = \text{Fraude Evitada} + \text{(-Custo da Campanha)} + \text{(-Fraude Perdida)}
$$</font>

* ***Resultado***: O valor l√≠quido que representa quanto  o modelo realmente "salvou" de dinheiro

```python
def simulacao_custo_fraude(y_true, y_pred, custo_fp = 1000, custo_fraude = 5000):
  """
  sa√≠da:
  -Tabela Principal: fp, fn, vp + linha total de  Impacto liquido
  -Tabela M√©tricas financeiras : ROI, taxa de Economia, Custo por RN

  """
  tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

  #Quantidades
  quantidade = [fp, fn, tp]

  #Custo da campanha(negativo para FP)
  custo_campanha = [-fp * custo_fp, 0, 0]

  #Fraude Evitada(vp)
  fraude_ev = [0, 0, tp * custo_fraude]

  #fraude n√£o Evitada(FN, negativo)
  fraude_perdida = [0, - fn * custo_fraude, 0]

  #Impacto l√≠quido = fraude evitada - custo da campanha - fraude n√£o evitada
  impacto_liquido = [fraude_ev[i] + custo_campanha[i] + fraude_perdida[i] for i in range(3)]

  #Tabela Principal
  df_principal = pd.DataFrame({
      "Quantidade": quantidade,
      "Custo Campanha(R$)": [f"R${x:,}" for x in custo_campanha],
      "Fraude Evitada(R$)": [f"R${x: ,}" for x in fraude_ev],
      "Fraude N√£o Evitada(R$)": [f"R${x: ,}" for x in fraude_perdida],
      "Impacto L√≠quido(R$)": [f"R$ {x: ,}" for x in impacto_liquido]
  }, index = ["FP", "FN", "VP"])

  #Linha Total
  total_quantidade = sum(quantidade)
  total_custo_campanha = sum(custo_campanha)
  total_fraude_ev = sum(fraude_ev)
  total_fraude_perdida = sum(fraude_perdida)
  total_impacto = sum(impacto_liquido)

  df_principal.loc["Total"] = [
      total_quantidade,
      f"R$ {total_custo_campanha:,}",
      f"R$ {total_fraude_ev:,}",
      f"R$ {total_fraude_perdida:,}",
      f"R$ {total_impacto:,}"
  ]

# M√©tricas financeiras derivadas
  custo_medio_fp = (fp * custo_fp / fp) if fp > 0 else 0
  roi = ((tp * custo_fraude) - (fp * custo_fp)) / (fp * custo_fp) if fp > 0 else None
  taxa_economia = (tp * custo_fraude) / ((tp * custo_fraude) + (fn * custo_fraude)) if (tp+fn)>0 else None
  custo_por_fn = (-fn * custo_fraude / fn) if fn > 0 else 0

  df_metricas = pd.DataFrame({
    "ROI (%)": [f"{roi*100:.1f}%" if roi is not None else "-"],
    "Taxa Economia (%)": [f"{taxa_economia*100:.1f}%" if taxa_economia is not None else "-"],
    "Custo por FN (R$)": [f"R$ {custo_por_fn:,}"]
}, index=["M√©tricas Financeiras"])

  return df_principal, df_metricas
```

## <font color = "green">Regress√£o Log√≠stica</font>

<font color = "darkpurple">**Regress√£o Losgictica**</font>: A regress√£o log√≠stica √© um modelo estat√≠stico usado para prever a probabilidade de um evento bin√°rio(sim/n√£o, verdadeiro/falso) acontecer, com base em uma ou mais vari√°veis independentes. Ela transforma a rela√ß√£o entre as vari√°veis em uma probabilidade entre 0 e 1 usando a fun√ß√£o logit, calculada pela f√≥rmula:

<font color = "darkpurple">$\text{logit}(p) = \log\left(\frac{p}{1-p}\right) = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_n X_n$</font>

onde *p* √© a probabilidade do evento, $X_1$ s√£o as vari√°veis independentes, e $\beta_1$ s√£o coeficientes ajustados via verossimilhan√ßa m√°xima para o melhor ajuste do modelo. O resultado final √© interpretado como chance de ocorr√™ncia do evento, e valores acima de 0,5 s√£o normalmente classificados como evento presente, e abaixo como evento ausente.

**Funcionamento:**

*  A regress√£o log√≠stica calcula a propabilidade do evento ocorrer, restrigindo essa probabilidade ao intervalo;

*   Usa a fun√ß√£o logit para linearizar a rela√ß√£o entre vari√°veis independentes e a chance do resultado;

*   Ajusta coeficientes por m√°xima verossimilhan√ßa para melhor previsibilidade;

*   Interpreta os coeficientes pelo "odds ratio", diminuindo aumento ou diminui√ß√£o da chance do evento.

**Problemas indicados para regress√£o log√≠stica:**

*   Classifica√ß√£o bin√°ria, como pr√©-doen√ßas(sim/n√£o), fraude financeira(fraude/n√£o fraude), risco de inadimpl√™ncia(alto/baixo);

*   Problemas onde se quer prever uma probabilidade categ√≥rica, especialmente com vari√°veis explicativas cont√≠nuas ou categ√≥ricas;

*  Aplica√ß√µes em setores como sa√∫de, finan√ßas, marketing, produ√ß√£o e epidemiologia.

## <font color = "green">Cross-Validation</font>

<font color = "darkpurple">StratifieldKFold</font>: √â uma t√©cnica de valida√ß√£o cruzada que preserva a propor√ß√£o original das classes do alvo em cada dobra do treinamento e teste. √â especialmente √∫til em problemas  de classifica√ß√£o com dados desequilibrados, garantindo que a distribui√ß√£o de classes minorit√°rias e majorit√°rias permane√ßa consistente em todos os subconjuntos, o que leva a medi√ß√µes de avalia√ß√£o mais confi√°veis.

**Funcionamento**

* O m√©todo divide o conjunto de dados em  **K** dobras(partes) de modo em que cada dobra contenha aproximadamente a mesma propor√ß√£o de exemplos de cada classe que existe no conjunto original;

* Em cada rodada, o modelo √© treinado em **K** - 1 dobras e testadas na dobra restante, como no KFold tradicional, mas preservando as propor√ß√µes das classes;

* Isso evita a situa√ß√£o em que alguma divis√£o resulta em um conjunto de treino ou teste sem exemplos da classe minorit√°ria, um problema comum em amostras examinadas simples.

**Exemplo pr√°tico**

Se um dataset tem 80% de uma classe e 20% de outra, cada dobra gerada pelo <font color = "darkpurple">StratifiedKFold</font> manter√° essas propor√ß√µes tanto no treino quanto no teste. Sen√£o usar a estratifica√ß√£o, pode ocorrer de um subconjunto conter quase apenas exemplos da classe majorit√°ria, prejudicando o aprendizado e a avalia√ß√£o do modelo.

**Quando Usar**

* Classifica√ß√£o bin√°ria ou multiclasse, especialmente em caso de desbalanceamento de dados;

* Sempre que a propor√ß√£o das aulas for importante para a robustez da avalia√ß√£o do modelo.

## <font color = "green">Voting Classifier</font>

<font color = "darkpurple">Voting Classifier</font>: O **Voting Classifier** √© um m√©todo de conjunto(ensemble) em aprendizado de m√°quina que combina a compreens√£o de m√∫ltiplos algoritmos base para gerar uma predi√ß√£o final mais robusta e muitas vezes mais precisa do que qualquer modelo individual. Ele funciona reunindo classifica√ß√µes diferentes(por exemplo, √°rvore de decis√£o, Svm, regress√£o log√≠stica) e tomando decis√µes por maioria de votos ou pela m√©dia das probabilidades de cada classe.

**Funcionamento**

* <font color = "purple">Hard Voting</font>: Cada classificador com base faz sua previs√£o e o **Voting Classifier** escolhe a classe mais votada entre eles(vota√ß√£o majorit√°ria). isso √© √∫til quando os modelos individuais tendem a funcionar bem em diferentes partes do conjuntos de dados;

* <font color = "purple"> Soft Voting</font>: Utiliza as probabilidades previstas para cada classificado. O modelo recomenda uma classe com maior m√©dia de probabilidades. Para isso, os modelos base precisam ter implementado o m√©todo predict_proba.

A combina√ß√£o dos modelos pode reduzir erros e melhorar a generaliza√ß√£o do sistema, especialmente quando os modelos t√™m desempenhos ou vis√µes diferentes. √â poss√≠vel tamb√©m calcular pesos aos classificados, dadno mais influ√™ncia aos modelos que historicamente t√™m melhores resultados.

**Quando Usar**:

* Quando voc√™ deseja unir diferentes tipos de modelos para aproveitar suas for√ßas complementares;

* Em c√©narios com alta variabilidade nos dados, onde diferentes algoritmos podem capturar diferentes padr√µes;

* Para aumentar a precis√£o, estabilidade e reduzir o overfitting.

## <font color = "green">Stacking Classifier</font>

<font color = "darkpurple">Stacking Classifier</font> O Stacking Classifier √© uma t√©cnica de conjunto em aprendizado de m√°quina que combina m√∫ltiplos algoratmos de classifica√ß√£o para formar um modelo ainda mais robusto e preciso. Diferentemente do Voting Classifier, o empilhamento utiliza um "meta-classificador" para aprender a combinar os modelos base e encontrar padr√µes que cada modelo individual n√£o consegue capturar sozinho.

**Funcionamento**

* Os modelos base(ex: √°rvores de decis√£o, regress√£o log√≠stica, KNN, SVM, etc) s√£o treinados normalmente no conjunto de dados originais;

* As desvios desses modelos para os dados de treinamento s√£o usadas como novos recursos para treinar um segundo modelo, chamado meta-classificador;

* O meta-classificador aprende a melhor forma de combina√ß√£o com a integra√ß√£o dos modelos base para gerar a previs√£o final;

*Normalmente,utiliza-se valida√ß√£o cruzada para gerar a varia√ß√£o dos modelos base, evitando overfitting.

**Benef√≠cios do empilhamento**

* Maior flexibilidade: pode combinar modelos muitos diferentes(√°rvores, redes, regress√£o, etc);

* Frquentemente aumenta o desempenho do sistema em compara√ß√£o com os modelos isolados, pois aproveita diferentes estrat√©gias de infer√™ncia;

* O metaclassificador pode aprender padr√µes,rela√ß√µes e erros que os modelos baseados individualmente n√£o detectam.

**Aplica√ß√µes**

* Competi√ß√£o de ci√™ncias de dados;

*Problemas complexos de classifica√ß√£o ou regress√£o com muitos padr√µes e variabilidade;

* Cen√°rios onde se busca extrair o m√°ximo do potencial de modelos diversos.

üìö<font color = "grey"> **Leituras Recomendadas**</font> :

[Bibliotecas Stacking Classifier](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.StackingClassifier.html)

<font color = "darkpurple">Considera√ß√µes da Aplica√ß√£o deste Modelo</font>:

Este modelo n√£o obteve um resultado satisfat√≥rio. Nas m√©tricas estat√≠sticas (PR AUC), o modelo s√≥ ficou √† frente dos modelos Random Forest e Isolation Forest. Tamb√©m n√£o se destacou nas demais m√©tricas. Nas m√©tricas financeiras, obteve um impacto l√≠quido de R$ 168.000,00 e um ROI de 121,4%, ou seja, ambos abaixo dos principais modelos, mas consumindo um poder computacional muito mais elevado. O fator decisivo para desconsiderar o modelo √© seu alto consumo de poder computacional, o que n√£o justifica os resultados inferiores.


```python
# --- 1. Selecionando as amostra dos dados ---
# Define o tamanho da amostra (no caso 10% dos dados originais)
sample_size = 0.08

# Selecionando uma amostra aleat√≥ria e estratificada dos dados de treino originais
X_train_sample, _, y_train_sample, _ = train_test_split(
    X_train_scaler, y_train, train_size=sample_size, random_state=70, stratify=y_train
)

# Selecionando uma amostra dos dados balanceados (SMOTE)
X_train_smote_sample, _, y_train_smote_sample, _ = train_test_split(
    X_train_smote, y_train_smote, train_size=sample_size, random_state=70, stratify=y_train_smote
)

print(f"Tamanho original do conjunto de treino: {len(X_train_scaler)}")
print(f"Tamanho da amostra de treino: {len(X_train_sample)}")
print("-" * 30)


# --- 2. Configurarando e instanciando os modelos ---
xgb_model_stack = XGBClassifier(
    n_estimators=80,
    learning_rate=0.2,
    max_depth=4,
    subsample=0.6,
    colsample_bytree=0.8,
    random_state=70,
    eval_metric="logloss"
)

gb_model_stack = GradientBoostingClassifier(
    n_estimators=80,
    learning_rate=0.2,
    max_depth=4,
    random_state=70,
)

meta_model = LogisticRegression(
    solver="lbfgs",
    max_iter= 800,
    random_state=70
)

stacking_clf = StackingClassifier(
    estimators=[
        ("XGBoost", xgb_model_stack),
        ("GradientBoosting", gb_model_stack)
    ],
    final_estimator=meta_model,
    cv=3,
    n_jobs=-1,
    passthrough=True
)


# --- 3. Cross-Validation e Treinamento com a amostra ---
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=70)

print("Iniciando a valida√ß√£o cruzada com a amostra...")
scores_stack = cross_val_score(
    stacking_clf,
    X_train_sample,
    y_train_sample,
    cv=cv,
    scoring="average_precision"
)

print("PR AUC m√©dio (CV) na amostra:", scores_stack.mean())
print("Desvio padr√£o na amostra:", scores_stack.std())
print("-" * 30)

print("Treinando o modelo final com a amostra SMOTE...")
stacking_clf.fit(X_train_smote_sample, y_train_smote_sample)
print("Modelo treinado com sucesso!")
print("-" * 30)


# --- 4. Predi√ß√£o e Avalia√ß√£o com os dados de teste completos ---
y_prob_stack = stacking_clf.predict_proba(X_test_scaler)[:, 1]

threshold = 0.4
y_pred_stack = (y_prob_stack >= threshold).astype(int)

# Chamando a fun√ß√£o de m√©tricas
metricas("Stacking Classifier", y_test, y_pred_stack, y_prob_stack)
```

## <font color = "green">Isolation Forest </font>

<font color = "darkpurple">Isolation Forest</font>: Isolation Forest √© um agoritmo de aprendizado de m√°quina n√£o supervisionado projetado principalmente para detec√ßao de anomalias(outliers) em conjuntos de dados, e √© muito eficiente para problemas com grandes volumes de dados e alta dimensionalidade. Inspirado em t√©cnicas de conjuntos baseados em √°rvores, ele opra diferente de outros m√©todos para isolar rapidamente inst√¢ncia que se desviam dos padr√µes normais.

**Funcionamento**

* O Isolation Forest construi v√°rias √°rvores de decis√£o de √°rvores de isolamento(iTrees), dividindo aleatoriamente os dados em cada ramo seletivo;

* A ideia central √© que anomalis s√£o poucas e diferentes, portanto s√£o isoaldas com menos divis√µes que inst√¢ncias normais;

* O algotitmo calcula um "tamanho de caminho" m√©dio para cada inst√¢ncia: anomalias tendem a ter caminhos mais curtos nas √°rvores, pois s√£o facilmente separadas dos dados;

* N√£o utiliza medi√ß√µes de densidade ou dist√¢nci, tornando-o r√°pido e escal√°vel.

**Principais aplica√ß√µes**

* Detec√ß√£o de fraudes banc√°rias;

* Monitoramento de falhas em sistemas ou fabrica√ß√£o;

* Identifica√ß√£o de outliers em exames m√©dicos;

* Qualquer cen√°rio em que dados espec√≠ficos/diferentes devem ser identificados em conjuntos grandes e complexos.

**Vantagens**

* Efici√™ncia: Isola anomalias rapidamente, mesmo em grandes volumes de dados;

* Menor consumo de mem√≥ria e processamento em compara√ß√£o com outros m√©todos;

* F√°cil implementa√ß√£o, especialmente com a biblioteca scikit-learn em Python

Nas m√©tricas financeiras, obteve um impacto l√≠quido negativo de R$ 309.000, um ROI de apenas 81,2% e uma fraca taxa de economia de 25,5%, a menor de todos os modelos. Isso aconteceu porque o modelo deixou passar o maior n√∫mero de fraudes entre todos os modelos testados

```python
#Instanciando o modelo n√£o supervisionado
iso = IsolationForest(
    n_estimators = 80,          # n√∫meros de arvores
    contamination = 0.0017,    # fra√ß√£o de fraudes esperada
    random_state = 70
)

#Treinando o modelos no dados
#neste modelo nao se treina com o y
iso.fit(X_train_scaler)

# Realizando as previs√µes
y_pred = iso.predict(X_test_scaler)

#converter: -1 anomalia -> 1 (fraude), 1 (normal) -> 0
y_pred = np.where(y_pred == -1, 1, 0)

#Matriz de confus√£o
print(f"Matriz de confus√£o:")
print(confusion_matrix(y_test, y_pred))

#Classification Report
print(f"\nClassification Report:")
print(classification_report(y_test, y_pred, digits = 4))

#Roc Auc
y_scores = iso.decision_function(X_test_scaler) * -1
roc_auc = roc_auc_score(y_test, y_scores)
print(f"\nROC AUC: {roc_auc:.4f}")

#PR AUC
pr_auc = average_precision_score(y_test, y_scores)
print(f"PR AUC : {pr_auc:.4f}")
```
